```python
import tensorflow as tf
from tensorflow.keras import layers
import pandas as pd
import numpy as np
from tensorflow.keras import datasets, layers, models
from tensorflow.keras.utils import to_categorical
```

```python
df = pd.read_csv(r"C:\Kodilla\18.3 neural network\iris.csv")
df.head()
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout
from keras.layers import Conv2D, MaxPooling2D
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(20, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
  ])
model

```

```python
from tensorflow import keras
from keras.callbacks import ReduceLROnPlateau
```

```python
# kompilujemy model
adam_opt = keras.optimizers.Adam(learning_rate=0.005)
model.compile(
    optimizer=adam_opt,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

```

```python
X = df.iloc[:,0:4].values
y = df.iloc[:,4].values
```

```python
Y = pd.get_dummies(y).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=4)
```

```python
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2,
                              patience=5, min_lr=0.001)

model.fit(X_train, y_train, batch_size=50, epochs=100, callbacks=[reduce_lr])
```

```python
y_pred = model.predict(X_test)
print("probs : ", y_pred[2])
print("actual :", np.argmax(y_pred,axis=1))
print("predict: ", np.argmax(y_test,axis=1))
```

```python
loss, accuracy= model.evaluate(X_test, y_test, verbose=0)
```

```python
print('Test loss:', loss)
print('Test accuracy:', accuracy)
```
